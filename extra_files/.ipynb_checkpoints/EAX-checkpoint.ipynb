```julia
neighbors = Dict{Int, Vector{Int}}()
for edge in solutions
    if edge[1] in keys(neighbors)
        push!(neighbors[edge[1]], edge[2])
    else
        neighbors[edge[1]] = [edge[2]]
    end

    if edge[2] in keys(neighbors)
        push!(neighbors[edge[2]], edge[1])
    else
        neighbors[edge[2]] = [edge[1]]
    end
end

neighbors
```

```julia
# Define a function to perform depth-first search
function dfs1(vertex::Int, prev::Int, neighbors::Dict{Int64, Vector{Int64}}, visited::Vector{Bool}, tour::Vector{Int})
    visited[vertex] = true
    for neighbor in neighbors[vertex]
        if neighbor != prev
            if neighbor > 0
                visited[neighbor] = true
                push!(tour, neighbor)
                return neighbor
            else
                return neighbor
            end
        end
    end
end

function dfs(vertex::Int, neighbors::Dict{Int64, Vector{Int64}}, visited::Vector{Bool}, tour::Vector{Int})
    visited[vertex] = true
    for neighbor in neighbors[vertex]
        if !visited[neighbor]
            dfs(neighbor, neighbors, visited, tour)
        end
    end
    push!(tour, vertex)
end

# Find all subtours using depth-first search
subtours = Vector{Vector{Int}}()
visited = fill(false, n)

for i=1:m
    vertex = pop!(neighbors[0])
    prev = 0
    tour = Int[0, vertex]
    while true
        temp = dfs1(vertex, prev, neighbors, visited, tour)
        prev = vertex
        vertex = temp
        if temp == 0
            deleteat!(neighbors[0], findfirst(x->x==prev, neighbors[0]))
            break
        end
    end
    push!(subtours, tour)
end

for vertex in [4,8,9,14]
    if !visited[vertex]
        tour = Int[]
        dfs(vertex, neighbors, visited, tour)
        push!(subtours, reverse(tour))
    end
    visited[1] = false
end
```
